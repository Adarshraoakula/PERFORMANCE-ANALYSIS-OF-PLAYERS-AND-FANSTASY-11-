```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
t20_data = pd.read_csv('t20.csv')
print(t20_data.shape)
# Basic information about the dataset
print("Basic Information:")
print(t20_data.info())

# Summary statistics for numeric columns
print("\nSummary Statistics:")
print(t20_data.describe())

# Check for null values
print("\nNull Values:")
print(t20_data.isnull().sum())

t20_data.drop(['Unnamed: 0', 'Unnamed: 15'], axis=1, inplace=True)


# Data Cleaning: Convert 'Runs' and '100' columns to numeric, setting errors='coerce' to handle non-numeric values
t20_data['Runs'] = pd.to_numeric(t20_data['Runs'], errors='coerce')
t20_data['100'] = pd.to_numeric(t20_data['100'], errors='coerce')

numeric_cols = ['Runs', '100']
for col in numeric_cols:
    t20_data[col].fillna(t20_data[col].mean(), inplace=True)

# Check for null values after handling
print("\nNull Values after Handling:")
print(t20_data.isnull().sum())


# Re-check the data types
print("\nData Types after Conversion:")
print(t20_data.dtypes)

# Visualizations
# Histogram of Runs
plt.figure(figsize=(10, 6))
sns.histplot(t20_data['Runs'], bins=30, kde=True)
plt.title('Distribution of Runs')
plt.xlabel('Runs')
plt.ylabel('Frequency')
plt.show()

# Boxplot of Runs vs Centuries
plt.figure(figsize=(10, 6))
sns.boxplot(x='100', y='Runs', data=t20_data)
plt.title('Runs Scored vs Centuries')
plt.xlabel('Centuries')
plt.ylabel('Runs')
plt.show()


import matplotlib.pyplot as plt

# Group data by the number of centuries and calculate the total runs for each group
grouped_data = t20_data.groupby('100')['Runs'].sum().reset_index()

# Plotting the bar graph
plt.figure(figsize=(12, 6))
plt.bar(grouped_data['100'], grouped_data['Runs'], color='skyblue', edgecolor='black')
plt.title('Total Runs Scored vs Number of Centuries')
plt.xlabel('Number of Centuries')
plt.ylabel('Total Runs Scored')
plt.xticks(grouped_data['100'])  # Ensure all century values are shown on the x-axis
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor

# Load the dataset
bowling_data = pd.read_csv('/Users/adarshraoakula/Downloads/602 project/Bowling_t20.csv')

# Cleaning and preparing the bowling data
bowling_data.drop(['Unnamed: 0', 'Span', 'Mat', 'Inns', 'Overs'], axis=1, inplace=True, errors='ignore')

# Convert Wkts and Ave to numeric, replacing non-convertible entries with NaN
bowling_data['Wkts'] = pd.to_numeric(bowling_data['Wkts'], errors='coerce')
bowling_data['Ave'] = pd.to_numeric(bowling_data['Ave'], errors='coerce')  # Correctly converting Ave to numeric

# Fill NaNs with the mean of the column
for column in bowling_data.columns:
    if bowling_data[column].dtype in ['float64', 'int64']:
        bowling_data[column].fillna(bowling_data[column].mean(), inplace=True)

# Define features and target for bowlers
X_bowl = bowling_data[['Wkts']]
y_bowl = bowling_data['Ave']

# Split the data into training and testing sets for bowlers
X_train_bowl, X_test_bowl, y_train_bowl, y_test_bowl = train_test_split(X_bowl, y_bowl, test_size=0.2, random_state=42)

# Decision Tree Regressor for bowlers
dt_bowl = DecisionTreeRegressor()
dt_bowl.fit(X_train_bowl, y_train_bowl)
bowling_data['predicted_ave'] = dt_bowl.predict(X_bowl)

# Sort bowlers based on highest Wkts and predicted averages
top_bowlers = bowling_data.sort_values(by=['Wkts', 'predicted_ave'], ascending=[False, True]).head(10)

# Display the results
print("Top 10 Bowlers based on Wickets taken and predicted averages:")
print(top_bowlers[['Player', 'Wkts', 'predicted_ave']])
```

```python
import pandas as pd

# Load the dataset
bowling_data = pd.read_csv('/Users/adarshraoakula/Downloads/602 project/Bowling_t20.csv')

# Ensure that the 'Mat' column is included if you need to reintroduce it
# bowling_data['Mat'] = ...

# Convert 'Ave' to numeric, replacing non-convertible entries with NaN
bowling_data['Ave'] = pd.to_numeric(bowling_data['Ave'], errors='coerce')
bowling_data['Mat'] = pd.to_numeric(bowling_data['Mat'], errors='coerce')

# Drop rows with NaN values in 'Ave' and 'Mat' to ensure accurate calculations
bowling_data.dropna(subset=['Ave', 'Mat'], inplace=True)

# Calculate the weighted average of the bowling average, weighted by the number of matches played
bowling_data['Weighted_Average_Ave'] = bowling_data['Ave'] * bowling_data['Mat']
total_weighted_ave = bowling_data['Weighted_Average_Ave'].sum()
total_matches = bowling_data['Mat'].sum()
weighted_average_ave = total_weighted_ave / total_matches

print("Weighted Average Bowling Average:", weighted_average_ave)
```
